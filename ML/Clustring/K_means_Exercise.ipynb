#Step 1 : Import the necessary Libraries
### import iris dataset


```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

#Step 2 : Load the iris Dataset

```python
dataset = load_iris()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
x = df.iloc[:,[0,1]].values

plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```python
x
```

#Step 3 : Elbow Method
Finding the ideal number of groups to divide the data into is a basic stage in any unsupervised algorithm. One of the most common techniques for figuring out this ideal value of k is the elbow approach.





```python
wcss = []
for i in range(1,11):
    Kmeans = KMeans(n_clusters=i,random_state=0)
    Kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
wcss = []
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

#Step 4 : Plot the Elbow graph to find the optimum number of cluster



```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

plotting WCSS and finding the point where adding more clusters yields only marginal improvemen

#Step 5 : Build the Kmeans clustering model



*   Find the cluster center
*   Predict the cluster group





```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(x)
```

#Step 6 : Predict the cluster group

```python
y_KMeans = kmeans.fit_predict(x)
plt.scatter(x[:,0],x[:, 1], c = y_KMeans)
plt.title('K-means clustring with k = 4')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```
